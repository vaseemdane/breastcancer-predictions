```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn import linear_model
import pandas as pd
from sklearn.metrics import accuracy_score
```

```python
cancer=datasets.load_breast_cancer()
```

```python
df=pd.DataFrame(cancer.data,cancer.target)
```

```python
df.head(5)
```

```python
df.columns=cancer.feature_names
print(df.columns)
```

```python
df.head(5)
```

```python
x=cancer.data
y=cancer.target
```

```python
print(f"the feature names {cancer.feature_names}")
```

```python
print(f"the targeted names {cancer.target_names}")
```

```python
df.shape
```

```python
x.shape
y.shape
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
l_reg=linear_model.LogisticRegression()
```

```python
model=l_reg.fit(x_train,y_train)
```

```python
predict=model.predict(x_test)
```

```python
acc=accuracy_score(y_test,predict)
```

```python
acc
```

```python
print("R^2: ", l_reg.score(x,y))
print("coeff: ", l_reg.coef_)
print("intercept: ", l_reg.intercept_)
```
